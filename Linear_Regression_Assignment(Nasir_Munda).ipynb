### Step 3: Data Preparation

```python
# Dummy variable month
months_df=pd.get_dummies(df.month,drop_first=True)
months_df.head()
```

```python
# Dummy variable weekday
weekdays_df=pd.get_dummies(df.weekday,drop_first=True)
weekdays_df.head()
```

```python
# Dummy variable weathersit
weathersit_df=pd.get_dummies(df.weathersit,drop_first=True)
weathersit_df.head()
```

```python
# Dummy variable season
seasons_df=pd.get_dummies(df.season,drop_first=True)
seasons_df.head()
```

```python
df.info()
```

```python
# Concatenating new dummy variables to the main dataframe
bike_sharing = pd.concat([df,months_df, weekdays_df, weathersit_df, seasons_df], axis = 1)
```

```python
bike_sharing.head()
```

```python
# Dropping columns season & weathersit since we have already created dummies for them
bike_sharing.drop(columns=["season", "month", "weekday", "weathersit"],axis=1,inplace =True)
```

```python
#Analysing dataframe after dropping columns
bike_sharing.info()
```

```python
#Plotting correlation heatmap to analyze the linearity between the variables in the dataframe 

plt.figure(figsize = (28, 20))
sns.heatmap(bike_sharing.corr(), annot = True, cmap="Greens")
plt.show()
```

```python
# splitting the dataframe into Train and Test
np.random.seed(0)
bike_sharing_train, bike_sharing_test = train_test_split(bike_sharing, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
# Shape of train data
bike_sharing_train.shape
```

```python
# Shape of test data
bike_sharing_test.shape
```

```python
# Assigning variable to scaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables

numeric_var = ['temp','atemp','humidity','windspeed','cnt']
bike_sharing_train[numeric_var] = scaler.fit_transform(bike_sharing_train[numeric_var])
```

```python
# Analysing the train dataframe after scaling
bike_sharing_train.head()
```

## Building the Linear Model

```python
y_train = bike_sharing_train.pop('cnt')
X_train = bike_sharing_train
```

```python
# Recursive feature elimination 

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(estimator=lm, n_features_to_select=15)
rfe = rfe.fit(X_train, y_train)
```

```python
# List of variables selected in top 15 list
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Creating X_test dataframe with RFE selected variables
col = X_train.columns[rfe.support_]
print(col)
```

```python
# checking which columns has been rejected

X_train.columns[~rfe.support_]
```

```python
# Generic function to calculate VIF of variables

def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
# Dataframe with RFE selected variables

X_train_rfe = X_train[col]
```

```python
# Calculate VIF
calculateVIF(X_train_rfe)
```

```python
# Building 1st linear regression model

X_train_lm_1 = sm.add_constant(X_train_rfe)
lr_1 = sm.OLS(y_train,X_train_lm_1).fit()
print(lr_1.summary())
```

```python
# As humidity shows high VIF values hence we can drop it
X_train_new = X_train_rfe.drop(['humidity'], axis = 1)

# Run the function to calculate VIF for the new model
calculateVIF(X_train_new)
```

```python
# Building 2nd linear regression model

X_train_lm_2 = sm.add_constant(X_train_new)
lr_2 = sm.OLS(y_train,X_train_lm_2).fit()
print(lr_2.summary())
```

```python
y_train_pred = lr_2.predict(X_train_lm_2)
```

```python
# Plot the histogram of the error terms

fig = plt.figure()
sns.distplot((y_train - y_train_pred), bins = 20)
fig.suptitle('Error Terms', fontsize = 20) 
plt.xlabel('Errors', fontsize = 18)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(X_train_new.corr(),annot = True, cmap="Greens")
plt.show()
```

```python
# Applying scaling on the test dataset

num_vars = ['temp', 'atemp', 'humidity', 'windspeed','cnt']
bike_sharing_test[num_vars] = scaler.transform(bike_sharing_test[num_vars])
bike_sharing_test.head()
```

```python
y_test = bike_sharing_test.pop('cnt')
X_test = bike_sharing_test
```

```python
col1 = X_train_new.columns

X_test = X_test[col1]

# Adding constant variable to test dataframe
X_test_lm_2 = sm.add_constant(X_test)
```

```python
y_pred = lr_2.predict(X_test_lm_2)
```

```python
r2 = r2_score(y_test, y_pred)
round(r2,4)
```

```python
# Visualizing the fit on the test data
# plotting a Regression plot

plt.figure()
sns.regplot(x=y_test, y=y_pred, ci=68, fit_reg=True,scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('y_test vs y_pred', fontsize=20)
plt.xlabel('y_test', fontsize=18)
plt.ylabel('y_pred', fontsize=16)
plt.show()
```
